```
import numpy as np
from matplotlib import pyplot as plt

n = 100

np.random.seed(42)
x = np.round(np.random.uniform(1,10, n), 3)
print ('X:', x)
```

```
import math

def calc_y(data):
    l = len(data)
    m = 0
    sd = 0.1
    # gaus_noise = np.random.normal(m, sd)
    np.random.seed(40)
    y = [np.log(x + np.random.normal(m, sd)) for x in data]
    return y

y = calc_y(x)
print ('Y:', y)

plt.scatter(x, y)
plt.xlabel('X-value')
plt.ylabel('Y-value')
plt.title('Scatter plot of the generated training samples')
plt.show()

def plot_neighbours(x,y, x_n, y_n, title="Actual vs Neighbours", y_label='Neighbours'):
    plt.scatter(x,y, c='b', marker='.', label='Actual')
    plt.scatter(x_n, y_n, c='r', marker='x', label=y_label)
    plt.legend(loc='upper left')
    plt.title(title)
    plt.show()
```

## Scheme One

```
test_set = [1, 3, 5, 7, 9]
k_set = [1, 3, 50]

def euc_dist(x1, x2):
    return np.sqrt((x1 - x2)**2)

for k in k_set:
    print ('*** For k =', k, '***')
    pred_list = []

    for val in test_set:
        print('*** For x-value:', val, '****')
        dist = [euc_dist(val, a) for a in x]
        indices = np.argpartition(dist, k)[:k]
        print ('indices of data with min distance:', indices)
        
        x_neigh = [x[i] for i in indices]
        y_neigh = [y[i] for i in indices]


        y_pred = sum(y_neigh)/k
        pred_list.append(y_pred)
        print (y_pred, '\n')

        plot_neighbours(x, y, x_neigh, y_neigh, "For k = " + str(k) + " and x = " + str(val) + ", Actual vs Neighbours")
    plot_neighbours(x, y, test_set, pred_list, "For k = " + str(k) + ", Actual vs Predicted", "Predicted")
    print('\n')
```

## Scheme Two

```
for k in k_set:
    print ('*** For k =', k, '***')
    pred_list = []

    for val in test_set:
        print('*** For x-value:', val, '****')
        dist = [euc_dist(val, a) for a in x]
        indices = np.argpartition(dist, k)[:k]
        print ('indices of data with min distance:', indices)

        w_list = [1/dist[i] for i in indices]
        y_list = [y[i] for i in indices]
        print ('weight:', w_list)
        print('y values of neighbours:', y_list)

        # normalizing the weights
        w_norm = [w/sum(w_list) for w in w_list]
        y_sum = 0

        for i in range(len(y_list)):
            y_sum += w_norm[i] * y_list[i]

        y_pred = y_sum
        pred_list.append(y_pred)
        
        print ('Predicted Y:', y_pred, '\n')
    plot_neighbours(x, y, test_set, pred_list, "For k = " + str(k) + ", Actual vs Predicted", "Predicted")
    print('\n')
```

## Scheme Three

```
def get_weight(d):
    return 1/np.exp(0.5*d*d)

y_pred_list = []
for val in test_set:
    print('*** For x-value:', val, '****')
    dist = [euc_dist(val, a) for a in x]

    # since every points contribute, therefore there are N neighbours
    w_list = [get_weight(d) for d in dist]
    print ('weight:', w_list)

    # normalizing the weights
    w_norm = [w/sum(w_list) for w in w_list]
    y_sum = 0

    for i in range(len(y_list)):
        y_sum += w_norm[i] * y[i]

    y_pred = y_sum
    y_pred_list.append(y_pred)
    print ('Predicted Y:', y_pred, '\n')
    print('\n')
plot_neighbours(x, y, test_set, y_pred_list, "Actual vs Predicted", "Predicted")
```
